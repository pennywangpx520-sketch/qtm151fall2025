(3) Recode a numeric column

<font size = "3">

- Import data from "races.csv" and assign it to a DataFrame
- Recode the "year" column into a new column "year_brackets"
with the following categories

$\qquad$ ``` ["1950-1965","1966-1980","1981-onwards"] ```

The easiest method is ``pd.cut()`` but you are not required to use this.

````python
# Your answer here
races = pd.read_csv("data_raw/races.csv")
bins = [1949, 1965, 1980, np.inf]                 
labels = ["1950-1965", "1966-1980", "1981-onwards"]
races['year_brackets'] = pd.cut(races['year'], bins=bins, labels=labels, include_lowest=True)

races.head()
````

Import the data from "sprint_results.csv" and assign it to a DataFrame (or use the same DataFrame you created in problem (2)). Do the following:

````python
# Your answer here
sprint = pd.read_csv("data_raw/sprint_results.csv")
df_avg_laps = (sprint
    .groupby('driverId')['laps']
    .mean()
    .sort_values(ascending=False)
)
df_avg_laps.head()
````
