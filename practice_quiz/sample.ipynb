```python
#1. if/elif/else: test whether a string belongs to list A or B
# name = input("Enter a string: ").strip()
name = "apple"

list_A = ["apple", "banana", "cherry"]
list_B = ["dog", "elephant", "fox"]

if name in list_A:
    print(f"'{name}' belongs to list A.")
elif name in list_B:
    print(f"'{name}' belongs to list B.")
else:
    print(f"'{name}' doesn't belong to either list.")
```

```python
#2. Build a NumPy array from two lists, then histogram it
import numpy as np
import matplotlib.pyplot as plt

# Example inputs
x_vals = [0.5, 1.2, -0.7, 2.1, 0.0, 3.4, -1.8]
y_vals = [0.0, 0.5, 1.0, 1.5, 3.14, 2.0, -0.5]

x = np.array(x_vals, dtype=float)
y = np.array(y_vals, dtype=float)

z = x**2 - 5 * np.sin(y)         # vectorized computation

plt.hist(z, bins=8, color="steelblue", edgecolor="black")
plt.title("Histogram of z = x^2 - 5 sin(y)")
plt.xlabel("z")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
#3. Two distributions  side-by-side histograms
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(42)
n = 2000

sample_normal = rng.normal(loc=0, scale=1, size=n)
sample_expon  = rng.exponential(scale=1.0, size=n)

fig, axes = plt.subplots(1, 2, figsize=(10, 4), sharey=True)

axes[0].hist(sample_normal, bins=30, edgecolor="black")
axes[0].set_title("Normal(0, 1)")
axes[0].set_xlabel("Value")
axes[0].set_ylabel("Count")

axes[1].hist(sample_expon, bins=30, edgecolor="black")
axes[1].set_title("Exponential(scale=1)")
axes[1].set_xlabel("Value")

fig.suptitle("Side-by-Side Histograms", y=1.03)
plt.tight_layout()
plt.show()
```

```python
#4. Read a CSV to a DataFrame and make a scatter plot

import pandas as pd
import matplotlib.pyplot as plt

us_emissions = pd.read_csv("practice_quiz_data/united_states_co2_emissions.csv")
print(us_emissions.columns)
plt.scatter(us_emissions["year"], us_emissions["total_emissions"], color="tomato", alpha=0.8)
plt.title("U.S. CO Emissions Over Time")
plt.xlabel("Year")
plt.ylabel("Total Emissions")
plt.tight_layout()
plt.show()

```

```python
#6. Loop over multiple CSVs, compute a column mean per file, store in a list

import pandas as pd
from pathlib import Path

files = {
    "United States": "practice_quiz_data/united_states_co2_emissions.csv",
    "Germany": "practice_quiz_data/germany_co2_emissions.csv",
}

target_col = "total_emissions"
means = []  # [(country, mean), ...]

for country, fname in files.items():
    df = pd.read_csv(fname)
    mean_val = df[target_col].mean(skipna=True)
    means.append((country, mean_val))
    print(f"{country}: {Path(fname).name}  mean {target_col}: {mean_val:.2f}")

print("\nCollected means:", means)

```
